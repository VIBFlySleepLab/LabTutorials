## Prepare circuit

The Servo should come with three cables bounded together. The colors are important:
    * Dark red goes to GND
    * Red goes to 5V
    * Orange goes to PIN 9

```python
from IPython.display import Image
from IPython.core.display import HTML 
Image(url= "https://programarfacil.com/wp-content/uploads/2015/09/conexion-servo-arduino.jpg")
```

```python
from pyfirmata import Arduino
import time
linux_port = "/dev/ttyACM0"
board = Arduino(linux_port)
```

```python
#!/usr/bin/python
# -*- coding: utf-8 -*-
# move a servo from a Tk slider - scruss 2012-10-28

import pyfirmata
from tkinter import *

# don't forget to change the serial port to suit
# board = pyfirmata.Arduino('/dev/tty.usbmodem26271')

# start an iterator thread so
# serial buffer doesn't overflow
iter8 = pyfirmata.util.Iterator(board)
iter8.start()

# set up pin D9 as Servo Output
pin9 = board.get_pin('d:9:s')

def move_servo(a):
    print(a)
    pin9.write(a)

# set up GUI
root = Tk()

# draw a nice big slider for servo position
scale = Scale(root,
    command = move_servo,
    to = 175,
    orient = HORIZONTAL,
    length = 400,
    label = 'Angle')
scale.pack(anchor = CENTER)

```

```python
# run Tk event loop
root.mainloop()
```
